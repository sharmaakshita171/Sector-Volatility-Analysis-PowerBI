```python
import yfinance as yf
import pandas as pd

tickers = ['XLK', 'XLV']
data = yf.download(tickers, start='2015-01-01', end='2025-01-01')['Close']
returns = data.pct_change().dropna()
rolling_vol = returns.rolling(window=30).std() * (252 ** 0.5)

df = pd.concat([data, returns, rolling_vol], axis=1, keys=['Price', 'Returns', 'Volatility'])
df.columns = [f'{a}_{b}' for a, b in df.columns]
df.dropna(inplace=True)
df.reset_index(inplace=True)

df.to_csv('sector_volatility.csv', index=False)                                                  
```
